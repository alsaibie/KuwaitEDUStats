```python
import pandas as pd
import numpy as np
import os
print(os.getcwd())
```

```python
#2018
df_2018_STEM_raw = pd.read_csv("sorted_data/2018/2018_STEM_raw.csv")
df_2018_Humanities_raw = pd.read_csv("sorted_data/2018/2018_Humanities_raw.csv")
df_2018_STEM_raw.head(10)
```

```python
# df_2018_STEM_raw.drop(df_2018_STEM_raw.columns[[]])
df_2018_STEM_fliplr = df_2018_STEM_raw[df_2018_STEM_raw.columns[::-1]]
df_2018_STEM_clean = df_2018_STEM_fliplr.drop(df_2018_STEM_fliplr.columns[[0, 1, 2, 8, 9]], axis=1)

df_2018_Humanities_fliplr = df_2018_Humanities_raw[df_2018_Humanities_raw.columns[::-1]]
df_2018_Humanities_clean = df_2018_Humanities_fliplr.drop(df_2018_STEM_fliplr.columns[[0, 1, 2]], axis=1)

display(df_2018_STEM_clean.head())
display(df_2018_STEM_clean.head())

df_2018_STEM_clean.to_csv("sorted_data/2018/2018_STEM_clean.csv", sep=",")
df_2018_Humanities_clean.to_csv("sorted_data/2018/2018_Humanities_clean.csv", sep=",")
```

```python
#2017
df_2017_Humanities_raw = pd.read_csv("sorted_data/2017/2017_Humanities_raw.csv")
display(df_2017_Humanities_raw.head())

df_2017_STEM_raw = pd.read_csv("sorted_data/2017/2017_STEM_raw.csv")
display(df_2017_STEM_raw.head())
```

```python
df_2017_STEM_fliplr = df_2017_STEM_raw[df_2017_STEM_raw.columns[::-1]]
df_2017_STEM_clean = df_2017_STEM_fliplr.drop(df_2017_STEM_fliplr.columns[[0,1]], axis=1)

df_2017_Humanities_fliplr = df_2017_Humanities_raw[df_2017_Humanities_raw.columns[::-1]]
df_2017_Humanities_clean = df_2017_Humanities_fliplr.drop(df_2017_Humanities_fliplr.columns[[0,1, 7]], axis=1)
df_2017_Humanities_clean = df_2017_Humanities_clean[np.isfinite(df_2017_Humanities_clean[u'النسبة'])]

display(df_2017_STEM_clean.head())
display(df_2017_Humanities_clean.head())

df_2017_STEM_clean.to_csv("sorted_data/2017/2017_STEM_clean.csv", sep=",")
df_2017_Humanities_clean.to_csv("sorted_data/2017/2017_Humanities_clean.csv", sep=",")

```
