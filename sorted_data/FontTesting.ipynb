```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager
flist = matplotlib.font_manager.get_fontconfig_fonts()
names = [matplotlib.font_manager.FontProperties(fname=fname).get_name() for fname in flist]
from bidi import algorithm as bidialg
import arabic_reshaper
%matplotlib inline

def arabic_text(utext):
    reshaped_text = arabic_reshaper.reshape(utext)
    return bidialg.get_display(reshaped_text)

text = arabic_text(u'كلام - وأرقام 2018')
yp = np.arange(0,200)

for k, family in enumerate(names):
    
    t = plt.text(0.8, yp[k], text + family, fontname=family)
```
